```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("emails.csv")
```

```python
data
```

```python
#eliminating 1st and last column and storing in x 
x=data.iloc[:,1:3001]
```

```python
x
```

```python
y=data.iloc[:,-1]
y
```

# Code for KNN

```python
import sklearn
```

```python
#splitting the dataset into 80% training data and 20% testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20)
```

```python
#data preprocessing
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
#fitting knn algo to training data
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=7)
classifier.fit(x_train,y_train)
```

```python
#predicting test result
y_pred=classifier.predict(x_test)
```

```python
#evaluation  of result using confusion matrix
from sklearn.metrics import classification_report, confusion_matrix

print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
#f1 score is harmonic mean of precision and recall and is
#better measure than accuracy
```

# Code for SVM


```python
from sklearn.svm import SVC
```

```python
#data preprocessing
#step 1: dividing data into attributes and labels
df=data.copy()
```

```python
df.dropna(inplace=True)
```

```python

X=df.drop(['Prediction'],axis=1)
Y=df['Prediction']
```

```python
#step 2:splitting into training and testing data
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
#training the SVM algorithm
svcclassifier=SVC(kernel='linear')
svcclassifier.fit(X_train,Y_train)
```

```python
#making predictions
Y_pred=svcclassifier.predict(X_test)
print(Y_pred)
```

```python
#evaluatung result of svm
print(classification_report(Y_test,Y_pred))
print(confusion_matrix(Y_test,Y_pred))
```
